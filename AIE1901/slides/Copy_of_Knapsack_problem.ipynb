Problem Instance:
You are a thief with a knapsack that can hold up to 15 kg. You've found 5 valuable items with the following weights and values:

Item	Weight (kg)	Value ($)
1	      2	         10
2	      5	         15
3	      8	         25
4	      3	         12
5	      6	         20
Objective: Select items to maximize total value without exceeding the 15 kg weight capacity.



```python
import coptpy as cp

################ Parameter #######################
weights = [2, 5, 8, 3, 6]  # weights of items
values = [10, 15, 25, 12, 20]  # values of items
capacity = 15  # knapsack capacity
n_items = 5 # you can only choose five items

# Create model
env = cp.Envr()
model = env.createModel("knapsack")

# Decision variables: x_i = 1 if item i is selected, 0 otherwise
x = model.addVars(n_items, vtype=cp.COPT.BINARY, nameprefix="x")

# Objective: maximize total value
model.setObjective(cp.quicksum(values[i] * x[i] for i in range(n_items)), cp.COPT.MAXIMIZE)

# Constraint: total weight <= capacity
model.addConstr(cp.quicksum(weights[i] * x[i] for i in range(n_items)) <= capacity)

# Solve the problem
model.solve()

# Print results
print("=== Knapsack Problem Solution ===")
print(f"Status: {model.status}")
print(f"Optimal objective value: ${model.ObjVal:.2f}")
print(f"Capacity used: {sum(weights[i] * x[i].x for i in range(n_items))} kg / {capacity} kg")
print("\nSelected items:")
for i in range(n_items):
    if x[i].x > 0.5:  # Check if item is selected
        print(f"Item {i+1}: weight={weights[i]}kg, value=${values[i]}")
```
